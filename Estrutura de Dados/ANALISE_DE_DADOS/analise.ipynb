Indentificar os meses com menor e maior numero de matriculas

```python
max_matr = df["Matrículas"].max()
print(max_matr)
mes_max_matr = df["Matrículas"].idxmax() 
mes_max_matr
```

```python
min_matr = df["Matrículas"].min()
print(min_matr)
mes_min_matr = df["Matrículas"].idxmin()
mes_min_matr
```

```python
print(f"O mês com maior número de matrículas é: {mes_max_matr}/mês: {max_matr} matrículas.")
print(f"O mês com menor número de matrículas é: {mes_min_matr}/mês: {min_matr} matrículas.")
```

```python
media_matr = df["Matrículas"].mean()
media_matr
```

Plot dessas informacoes

```python
plt.figure(figsize=(10,5)) # tamanho do gráfico
df["Matrículas"].plot(marker='o',linestyle='-',color="blue",label="Matrículas", mfc='black') # definiu coluna principal
# marker - delimitador, 'o' vai ser redondo, mfc - altera a cor dela
# label - é o titulo 
plt.grid(True)

# adicionar linha de info min e max de matricula 
# máximo
plt.axhline(y=max_matr,color="yellow",linestyle='--',linewidth=2, label=f"Máximo: {max_matr}") # tracejado horizontal 
# linewidth - espressura da linha
# label - legenda

# minimo
plt.axhline(y=min_matr,color="red",linestyle="--",linewidth=2,label=f"Mínimo: {min_matr}")

# adicionar média 
plt.axhline(y=media_matr,color="black",linestyle="--",linewidth=2,label=f"Média: {media_matr:.2f}")

# para aparecer a legenda(label)
plt.legend()

# definir titulos e legendas
plt.title('Matrículas mensais')
plt.xlabel("Mês")
plt.ylabel("Números de matrículas")
```

Tendencia de Matrículas

Calcular a media de matriculas por trimestre 

QE - Trimestre
QS - Semestre
Q  - Quadrimestre
A  - Anual
M  - Mensal
B  - Diário
W  - Semanal
H  - Horárioa

```python
# Passo 1 - Converter pra Datetime
df["Mês"] = pd.to_datetime(df["Mês"])

# Passo 2 - Definir a coluna de datas como indices
df = df.set_index("Mês")

# Passo 3 - Calcular a media trimestral
matr_tri = df["Matrículas"].resample("QE").mean()
# resample - define o calculo, usando as siglas acima
#exemplo qe-trimestre
print(matr_tri)
```

Gráfico de barras

```python
plt.figure(figsize=(10,5))
matr_tri.plot(kind='bar',color='blue')
# kind - tipo de gráfico    bar - barras
plt.title('Média de Matrículas por Trimestre')
plt.xlabel("Trimestre")
plt.ylabel("Média de Matrículas")
plt.xticks(rotation=45) # quanto a rotacao do eixo x , das palavras
plt.grid(axis='y') # a linha do gráfico, escolhe qual eixo vai estar, True - ela fica horizontal e vertical
```
